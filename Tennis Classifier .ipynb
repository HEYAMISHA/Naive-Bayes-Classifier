#  Method 1 - Prediction Using Bayes' Theorem :

```python
def pro(): # For finding likelihood of 'Yes' given condition of the feautures
    feature=input("Enter the feature = ")
    cond=input("Enter the condition of the feature = ")
    ct1 = pd.crosstab(data[feature], data['Play Tennis'],colnames=["Play Tennis"],rownames=[feature],margins = True,margins_name="Total",normalize='columns')
    ct2 = pd.crosstab(data[feature], data['Play Tennis'],colnames=["Play Tennis"],rownames=[feature],margins = True,margins_name="Total",normalize='index')
    return ct1.loc[cond,"Yes"]*ct2.loc["Total","Yes"]/ct1.loc[cond,"Total"]
for i in range(4):
    Prob_Yes=1
    Prob_No=1
    Yes_cond=pro()
    No_cond=1 - Yes_cond
    Prob_Yes=Prob_Yes*Yes_cond
    Prob_No=Prob_No*No_cond
print("Probability of 'Yes' on that day = ",Prob_Yes,"\nProbability of 'No' on that day = ",Prob_No,"\n")    
if Prob_Yes > Prob_No:
    print("PREDICTION: You can play tennis on that day.")
else:
    print("PREDICTION: You cannot play tennis on that day.")
```

#  Method 2 - Prediction Using Naive Bayes With Sckit-learn :

```python
import numpy as np
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
#Import scikit-learn metrics module for accuracy calculation
from sklearn.metrics import accuracy_score
number = LabelEncoder() # Encodes alphabetically
data['Outlook'] = number.fit_transform(data['Outlook'])
data['Temperature'] = number.fit_transform(data['Temperature'])
data['Humidity'] = number.fit_transform(data['Humidity'])
data['Wind'] = number.fit_transform(data['Wind'])
data['Play Tennis'] = number.fit_transform(data['Play Tennis'])
features = ["Outlook", "Temperature", "Humidity", "Wind"]
target = "Play Tennis"
features_train, features_test, target_train, target_test = train_test_split(data[features],data[target],test_size = 0.33,random_state = 54)

#Create a Gaussian Classifier
model = GaussianNB()
# Train the model using the training sets
model.fit(features_train, target_train)
#Predict the response for test dataset
pred = model.predict(features_test)
print('Tennis data after encoding the features =\n ',data)
# Model Accuracy, how often is the classifier correct?
accuracy = accuracy_score(target_test, pred)
print('Accuracy of the Naive Bayes Classifier = ',accuracy)
#Predict Output
print('Predicted value for the given features:',model.predict([[1,1,1,0]]))
```
